## 이것을 어떻게 학습할 수 있을까??

#### 결론적으로 백프로파게이션이다.

Y(결과값)으로 Cost 값을 구하고, 경사하강법을 통해서 최소 cost를 구한다.

경사하강법을 구하기 위해서는 결국 미분을 계산해야한다. 그럼 여기서 미분을 어떻게 구할까??

간단한 예제를 정의하고 백프로파게이션을 이용해 미분을 구해보자


image.png

f = wx + b를 위와같은 그림으로 나타낼 수 있다. 그리고 보라색 글씨는 각 w, x, b가 f에 미치는 영향이다. 즉, 이게 미분값(기울기)이다.

편미분의 원리를 따라서 f를 w,x로 미분하기위해 g로 미분해서 인자를 통해서 미분값을 구할수있다. 

이는 위 그래프가 여러개 연결되어도, 편미분의 인자들을 통해서 연속적으로 미분값을 구할 수 있는것이다. 이를 TensorFlow에서 아래 식으로 구현하여 만들어 두었다.
