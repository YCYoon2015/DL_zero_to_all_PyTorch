```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
# For reproducibility
torch.manual_seed(1)
```

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[1], [2], [3]])

W = torch.zeros(1) #모델 초기화

lr = 0.1

nb_epochs = 10
for epoch in range(nb_epochs + 1):
  hypothesis = x_train * W

  cost = torch.mean((hypothesis - y_train) ** 2) #cost 계산
  gradient = torch.mean((W * x_train - y_train) * x_train) #cost gradient 계산

  print('Epoch {:4d}/{} W: {:.3f}, Cost: {:.6f}'.format(
      epoch, nb_epochs, W.item(), cost.item()
  ))

  W -= lr * gradient # cost gradient로 H(x) 개선
```

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[100], [200], [300]])

W = torch.zeros(1, requires_grad=True)

optimizer = optim.SGD([W], lr=0.15)

nb_epochs = 10
for epoch in range(nb_epochs + 1):

  hypothesis = x_train * W

  cost = torch.mean((hypothesis - y_train)**2)

  print(f'Epoch {epoch:4d}/{nb_epochs} W: {W.item():.3f} Cost: {cost.item():.6f}')

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
```
