### selection & drop

```python
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data'
df_data = pd.read_csv(data_url, sep='\s+', header = None) # 주소, 나누는기준(띄어쓰기 여러개), 해더없음
```

```python
df_data
```

```python
df_data.head(2).T
```

```python
# 데이터 뽑기
df_data[[0, 1]] # column명을 list에 적어서 df으로 뽑을 수 있다.
```

```python
df_data[[0, 1]][:5] # index도 가능하다
```

```python
# location을 이용하면 순서가 반대이다.
df_data.loc[[0,5], [0, 1]] # index, column명
```

```python
# iloc은 index로 다룬다.
df_data.iloc[:5,:2] # index, column의 index
```

### index 다루기

```python
df_data
```

```python
df_data.reset_index() # 새롭게 index row 생성
```

```python
df_data.reset_index(drop=True) # 새롭게 index 생성 및 기존 index 삭제
```

```python
df_data.reset_index(inplace=True) # inplace는 df를 변경하는 명령임 (최소화 해야함)
```

```python
df_data
```

### data drop

```python
df_data.drop(1) # idx = 1의 row를 삭제 (영구반영하려면 inplace=True)
```

```python
df_data.drop([1,4,5]) # idx = 1,4,5의 row를 삭제 (영구반영하려면 inplace=True)
```

```python
df_data.drop([1,4], axis=1) # 컬럼 삭제
```

```python
df_data.drop(['index'], axis=1, inplace=True) # 컬럼 삭제
```

```python
df_data
```
